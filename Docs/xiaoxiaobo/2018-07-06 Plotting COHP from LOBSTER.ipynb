```python
from  pymatgen.electronic_structure.cohp  import Cohp
from pymatgen.electronic_structure.plotter import CohpPlotter
from pymatgen.io.lobster import Cohpcar
#测试是否能成功
COHPCAR_path = "E:/Python/pymatgen-master/test_files/cohp/COHPCAR.lobster"
cohpcar = Cohpcar(filename=COHPCAR_path)
#获取COHPCAR文件的名称
cdata = cohpcar.cohp_data
cdata_processed = {}
for key in cdata:
    #PYTHON里的for循环
    c = cdata[key]
    c["efermi"] = 0
    c["energies"] = cohpcar.energies
    c["are_coops"] = False
    cdata_processed[key] = Cohp.from_dict(c)
    #返回完整的Cohp对象
cp = CohpPlotter()
cp.add_cohp_dict(cdata_processed)
x = cp.get_plot()
x.ylim([-6, 6])
x.show()
#结果展示
```

这个例子是关于如何使用局部—轨道基础套件电子结构重建（LOBSTER）来绘制计算汉密尔顿总体（COHP）的。COHP将能带结构能量划分为轨道对相互作用，从化学角度来说，它是一对相邻原子之间的“键加权”状态密度。换句话说COHP暗示了粘合强度，它以每摩尔ev或千焦耳为单位。而LOBSTER就是计算COHP的，它消除了VASP,ABINIT和Quantum ESPRESSO的电子结构输入。

for key in cdata:是PYTHON里的循环语句，把cdata的值依次赋值给Key.
Get-plot（）获取显示COHP的matplotlib图，xlim-指定x轴限制，默认为自动适应，ylim-指定y轴限制，默认为自动适应
